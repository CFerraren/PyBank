### Manual Differencing

```python
# create a differenced series
def difference(dataset, interval=12):
    diff = list()
    for i in range(interval, len(dataset)):
        value = dataset[i] - dataset[i - interval]
        diff.append(value)
    return pd.DataFrame(diff, columns=['rvn_diff'])
```

```python
X = df.Revenue.values
X = X.astype('float32')
```

```python
# difference data
months_in_year = 12
stationary = difference(X, months_in_year)
stationary.index = df.index[months_in_year:]
```

```python
ax=sns.distplot(stationary, color='#2ecc71')
```

#### L2 Normalization to Scale Data 

```python
stationary['rvn_normalized'] = preproc.normalize(stationary[['rvn_diff']], axis=0)
```

```python
sns.distplot(stationary.rvn_normalized, color='#2ecc71')
```

```python
# check if stationary
result = adfuller(stationary.rvn_normalized)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
rvn = stationary.rvn_normalized
```

```python
rvn.rolling(12).mean().plot(linewidth=1, fontsize=12, color='#2ecc71')
plt.show()
```

### Seasonability and Autocorrelation

```python
#Using 1st discrete difference of object

rvn.diff(periods=4).plot(linewidth=1, fontsize=12,  color='#2ecc71')
plt.show()
```

```python
#correlogram
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(rvn)
plt.show()
```

```python
#correlogram
plot_acf(rvn, lags=50)
plt.show()
```

```python
#partial autocorrelation
plot_pacf(rvn, lags=50)
plt.show()
```

```python
#import arima library
from statsmodels.tsa.arima_model import ARIMA
```

```python
#model and fit data to the model
model = ARIMA(rvn, order=(10,1,1))
model_fit = model.fit(disp=0)
```

```python
print(model_fit.summary())
```

```python
# plot residual errors
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
plt.show()
residuals.plot(kind='kde')
plt.show()
print(residuals.describe())
```

```python
X=rvn.values
X[:5]
```

```python
size = int(len(X) * 0.5)
```

```python
train, test = X[0:size], X[size:len(X)]
```

```python
history = [x for x in train]
predictions = list()
```

```python
for t in range(len(test)):
    model = ARIMA(history, order=(10,1,1))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
```

```python
len(test), len(predictions)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
```

```python
# plot
plt.plot(test)
plt.plot(predictions, color='red')
plt.show()
```
