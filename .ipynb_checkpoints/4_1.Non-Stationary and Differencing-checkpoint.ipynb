#### Appy Power Transform - Yeo_Johnson 

````python
modeler = preproc.PowerTransformer(method='yeo-johnson')
modeler.fit(x)
df['rvn_transformed'] = modeler.fit_transform(x)
x_yeo= df[['rvn_transformed']]
````

````python
#plot the distribution
ax=sns.distplot(x_yeo, color='#2ecc71')
ax.tick_params(labelsize=10)
ax.set_xlabel('After Yeo-Johnson Power Transformed Revenue', fontsize=10)
ax.set_ylabel('Freq', fontsize=10)
plt.show()
````

- We now have a Gaussian distribution after `Yeo-Johnson` transformed.

### Seasonality

#### Check Data Seasonality by Summary Statistics

````python
#check for seasonality
def stationarity(X):
    split = int(len(X) / 2)
    X1, X2 = X[0:split], X[split:]
    mean1, mean2 = X1.mean(), X2.mean()
    var1, var2 = X1.var(), X2.var()
    print('mean1=%f, mean2=%f' % (mean1, mean2))
    print('variance1=%f, variance2=%f' % (var1, var2))
````

````python
# Before ``Yeo-Johnson`` Transform
stationarity(x)
````

````python
#After the ``Yeo-Johnson`` PowerTransform
stationarity(x_yeo)
````

- Mean absolute values are the same. Our data is stationary after we have applied the earlier the Yeo-Johnson transform to make our distribution more Gaussian and linear.

#### Check Using Augmented Dickey-Fuller

- Type of statistical test called a unit root test


- Unit root test is that it determines how strongly a time series is defined by a trend.

````python
x_yeo_val = x_yeo.values
x_yeo_val= x_yeo_val.flatten()
x_yeo_val[:5]
````

````python
a_result = adfuller(x_yeo_val)
print('ADF Statistic: %f' % a_result[0])
print('p-value: %f' % a_result[1])
print('Critical Values:')
for key, value in a_result[4].items():
    print('\t%s: %.3f' % (key, value))
````

- We got second confirmation that our data is stationary. The null-hypothesis is rejected with a significance level of less than 1%; . ADF statistics is -9.13 and is less than 1%, meaning our data is not a statistical fluke.


- It alo means that our data has no unit root or does not have time-dependent structure.

#### Differencing

- Used to remove temporal dependence like trends and seasonality.

````python
# create a differenced series
def difference(dataset, interval=1):
    diff = list()
    for i in range(interval, len(dataset)):
        value = dataset[i] - dataset[i - interval]
        diff.append(value)
    return pd.Series(diff)
````

````python
# invert difference
def invert_difference(orig_data, diff_data, interval):
    orig = list()
    for i in range(interval, len(orig_data)):
        value = diff_data[i-interval] + orig_data[i-interval]
        orig.append(value)
    return pd.Series(orig)
````

````python
X = x_yeo.rvn_transformed.values
X = X.astype('float32')
````

````python
# difference data
months_in_year = 12
lag_diff = difference(X, months_in_year)
lag_diff.index = x_yeo.index[months_in_year:]
````

````python
# plot
plt.plot(lag_diff, color='#2ecc71')
plt.show()
````

````python
# Use decoposition to check for trend and seasonality
decomposition_lag = sm.tsa.seasonal_decompose(lag_diff, model='additive')
fig = decomposition_lag.plot()
plt.show()
````

- After detrend, the upward-trend is stabilized horizontally. 

````python
# prepare dataset
train_size = int(len(lag_diff) * 0.66)
train, test = lag_diff[0:train_size], lag_diff[train_size:]
````

````python
# persistence
predictions = list()
history = train[-1]
for i in range(len(test)):
    yhat = history
    predictions.append(yhat)
    history = test[i]
error = mean_squared_error(test, predictions)
print('Persistence MSE: %.3f' % error)
````

````python
# random walk prediction
predictions_rw = list()
history_rw = train[-1]
for i in range(len(test)):
    yhat = history_rw + (-1 if random() < 0.5 else 1)
    predictions_rw.append(yhat)
    history_rw = test[i]
error = mean_squared_error(test, predictions_rw)
print('Random MSE: %.3f' % error)
````
