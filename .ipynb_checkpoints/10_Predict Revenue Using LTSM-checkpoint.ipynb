# Use Long Short Term Memory (LSTM) to Predict Revenue

In this project, I created a Python notebook for predicting revenue of Bank using Long short Term Memeory which are widely used for sequence prediction problems and have proven to be extremely effective. 

LTSM has three gates:

- **The input gate**: The input gate adds information to the cell state

- **The forget gate**: This removes information that are not required by the model 

- **The output gate**: Select the information to be shown as output

```python
#import packages
import numpy as np
import pandas as pd
import os
```

```python
#use seaborn to prettify the plots
import seaborn as sns
sns.set_style('whitegrid')
```

```python
#display pd typeformat
pd.set_option('display.float_format', '{:,.0f}'.format)
```

```python
#directory and filename
dir = 'data/processed/'
file = 'revenue_2009_2016.csv'
```

```python
#load csv into dataframe and print the head
#parse the date colum to datetime format
#set 'Date' column to index
df = pd.read_csv(os.path.join(dir, file), parse_dates=['Date'])
df[:5]
```

```python
df.describe().T
```

```python
#make a copy
new_data = df.copy()
new_data = new_data.sort_index(ascending=True, axis=0)
new_data.index = new_data['Date']
new_data[:5]
```

```python
#plot
plt.plot(new_data['Revenue'], label='Revenue')
```

```python
new_data[:5]
```

```python
#drop date column
new_data.drop('Date', axis=1, inplace=True)
new_data[:5]
```

```python
#creating train and test sets
dataset = new_data.values
```

```python
train = dataset[0:78,:]
valid = dataset[78:,:]
```

```python
#for normalizing data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
#converting dataset into x_train and y_train
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dataset)
```

```python
x_train, y_train = [], []

for i in range(1,len(train)):
    x_train.append(scaled_data[i-1:i,0])
    y_train.append(scaled_data[i,0])
    
x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1],1))
```

```python
#importing required libraries
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
```

```python
# create and fit the LSTM network
model = Sequential()
model.add(LSTM(units=1000, return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(LSTM(units=1000))
model.add(Dense(1))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(x_train, y_train, epochs=1, batch_size=1, verbose=2)
```

```python
#predicting values, using past data from the train data
inputs = new_data[len(new_data) - len(valid) - 1:].values
inputs = inputs.reshape(-1,1)
inputs  = scaler.transform(inputs)
```

```python
X_test = []
for i in range(1,inputs.shape[0]):
    X_test.append(inputs[i-1:i,0])
    
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))

pred_rvn = model.predict(X_test)
pred_rvn = scaler.inverse_transform(pred_rvn)
```

```python
rms=np.sqrt(np.mean(np.power((valid-pred_rvn),2)))
rms
```

```python
#for plotting
train = new_data[:78]
valid = new_data[78:]
valid['Predictions'] = pred_rvn
plt.plot(train['Revenue'])
plt.plot(valid[['Revenue','Predictions']])
```
