### Import Libraries

```python
#import packages
import numpy as np
import pandas as pd
from scipy import stats
import os
import sklearn.preprocessing as preproc
from statsmodels.tsa.stattools import adfuller
from astropy.table import Table

#to plot within the notebook
import matplotlib.pyplot as plt
%matplotlib inline

#use seaborn to prettify the plots
import seaborn as sns
sns.set_style('whitegrid')
```

```python
#setting figure size
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 20,10
```

```python
#display pd typeformat
pd.set_option('display.float_format', '{:,.0f}'.format)
```

### Load Data

```python
#directory and filename
dir = 'data/processed/'
file = 'revenue_2009_2016.csv'
```

```python
#load csv into dataframe and print the head
#parse the date colum to datetime format
#set 'Date' column to index
data = pd.read_csv(os.path.join(dir, file), parse_dates=['Date'], index_col='Date')
```

```python
#make a copy
df = data.copy()
```

```python
x = df[['Revenue']]
```

```python
#plot the distribution
ax=sns.distplot(x, color='#2ecc71')
```

```python
#check for stationarity
def stationarity(X):
    split = int(len(X) / 2)
    X1, X2 = X[0:split], X[split:]
    mean1, mean2 = X1.mean(), X2.mean()
    var1, var2 = X1.var(), X2.var()
    print('mean1=%f, mean2=%f' % (mean1, mean2))
    print('variance1=%f, variance2=%f' % (var1, var2))
```

### Data Transformation

```python
stationarity(x)
```

> Our revenue data is non-stationary; its absolute value of means must be equally close.

---

#### PowerTransform

```python
# PowerTransform  - some outputs will be negative
df['PowerTransform']= preproc.power_transform(x,method='yeo-johnson')
```

```python
stationarity(df['PowerTransform'])
```

> After Yeo-Johnson Powertransform, both means now are absolute equal.

---

#### Using StandardScaler

- Standardization is a transform for data with a Gaussian distribution.

```python
# Standardization  - some outputs will be negative
df['standardized'] = preproc.StandardScaler().fit_transform(x)
```

```python
stationarity(df['standardized'])
```

> After Standardization transformed, both means now are absolute equal.

---
#### Min-max Scaling

- Normalization is a rescaling of data from the original range to a new range between 0 and 1. 

- scikit-learn MinMaxScaler class can be used to rescale data to any range you wish by specifying the preferred range in the constructor of the object.

```python
# Min-max scaling
df['minmax'] = preproc.minmax_scale(x)
```

```python
stationarity(df['minmax'])
```

 ---

#### L2 Normalization

> Same with Min-max Scaling.

```python
df['l2_normalized'] = preproc.normalize(x, axis=0)
```

```python
stationarity(df['l2_normalized'])
```

---
#### Plot

```python
# Visualizing the histograms of original and scaled data
plt.figure(figsize=(10,20))

# original review count histogram
ax = plt.subplot(5,1,1)
sns.distplot(x, color='#2ecc71')
ax.tick_params(labelsize=10)
ax.set_xlabel('Bank Revenue', fontsize=10)
ax.set_ylabel('Freq', fontsize=10)

# Scaled using the Power Transform
ax = plt.subplot(5,1,2)
sns.distplot(df['PowerTransform'], color='#2ecc71')
ax.tick_params(labelsize=10)
ax.set_xlabel('Power Transformed Revenue', fontsize=10)
ax.set_ylabel('Freq', fontsize=10)

# Scaled using the Variance Scaling or Standardization
ax = plt.subplot(5,1,3)
sns.distplot(df['standardized'], color='#2ecc71')
ax.tick_params(labelsize=10)
ax.set_xlabel('Variance Scaled Revenue', fontsize=10)
ax.set_ylabel('Freq', fontsize=10)

# Scaled using min-max scaling
ax = plt.subplot(5,1,4)
sns.distplot(df['minmax'], color='#2ecc71')
ax.tick_params(labelsize=10)
ax.set_xlabel('Min-Max Scaled Revenue', fontsize=10)
ax.set_ylabel('Freq', fontsize=10)


# Scaled using the L2 Normalization
ax = plt.subplot(5,1,5)
sns.distplot(df['l2_normalized'], color='#2ecc71')
ax.tick_params(labelsize=10)
ax.set_xlabel('L2 Normalized Revenue', fontsize=10)
ax.set_ylabel('Freq', fontsize=10)

plt.show()
```
