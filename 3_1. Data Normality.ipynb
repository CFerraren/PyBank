### Load Data

```python
#directory and filename
dir = 'data/processed/'
file = 'revenue_2009_2016.csv'
```

```python
#load csv into dataframe and print the head
#parse the date colum to datetime format
#set 'Date' column to index
data = pd.read_csv(os.path.join(dir, file), parse_dates=['Date'], index_col='Date')
```

```python
#make a copy
df = data.copy()
```

```python
x = df['Revenue']
```

```python
#plot the distribution
ax=sns.distplot(x, color='#2ecc71')
```

```python
#check for normality
def normtesttab(x):
    nm_stat, p_norm = stats.normaltest(x)
    jb_stat, p_jar = stats.jarque_bera(x)
    data_rows = [('D’Agostino-Pearson', nm_stat, p_norm),
                 ('Jarque-Bera', jb_stat, p_jar)]
    t = Table(rows=data_rows, names=('Test name', 'Statistic', 'p-value'), 
              meta={'name': 'normal test table'},
          dtype=('S25', 'f8', 'f8'))
    print(t)
```

```python
normtesttab(x)
```

```python
#Compute optimal Yeo-Johnson transform parameter for input data, 
#using maximum likelihood estimation

lmax = stats.yeojohnson_normmax(x)
fig = plt.figure()
ax = fig.add_subplot(111)
prob = stats.yeojohnson_normplot(x, -20, 20, plot=ax)
ax.axvline(lmax, color='r')
```

```python
#transform uisng yeojohnson
xt, maxlog = stats.yeojohnson(x)
print("lambda = {:g}".format(maxlog))
```

```python
normtesttab(xt)
```

```python
#Plot both after applying "Yeojohnson" Power Transform
x1=x
x1=(x1 - x1.mean()) / (x1.max() - x1.min())
 
xt1=xt
xt1=(xt1 - xt1.mean()) / (xt1.max() - xt1.min())
 
sns.kdeplot(x1, shade=True, label="Raw (normalized)" );
sns.kdeplot(xt1, shade=True, label="Transformed (normalized)" )
```
